```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#TASK 1
```

```python
df = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
df.head()
```

```python
df['genres'] = df['genres'].apply(lambda x: [item['name'] for item in json.loads(x)])
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: [item['name'] for item in json.loads(x)])
```

```python
df.shape
```

```python
df[['title', 'genres']].head(50)
```

```python
#TASK 2
```

```python
df.isnull().sum()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
object_columns = df.select_dtypes(include=['object']).columns

# Impute numeric columns with mean
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Impute object columns with a placeholder (you can change this placeholder as needed)
df[object_columns] = df[object_columns].fillna('Not Available')
```

```python
#TASK 3
```

```python
f_df = df[df['budget'] > 220000]
genres_highbudget = f_df['genres']

# Display movie categories (genres) with a budget greater than $220,000
print(genres_highbudget)
```

```python
#TASK 4
```

```python
f_df1 = df[df['revenue'] > 961000000]
genres_highrev = f_df1['genres']
print(genres_highrev)
```

```python
#TASK 5
```

```python
df = df[(df['budget'] != 0) & (df['revenue'] != 0)]
print(f"Number of rows after removing rows with 0 budget and revenue: {df.shape[0]}")
```

```python
#TASK 6
```

```python
# Top 10 movies with the highest revenue
top_10_high_revenue = df.nlargest(10, 'revenue')[['title', 'revenue']]
print("Top 10 movies with the highest revenue:")
print(top_10_high_revenue)

# Top 10 movies with the least budget
top_10_low_budget = df.nsmallest(10, 'budget')[['title', 'budget']]
print("\nTop 10 movies with the least budget:")
print(top_10_low_budget)
```

```python
#TASK 10
```

```python
# Sort data in descending order of revenue
sorted_data = df.sort_values(by='revenue', ascending=False)

# Filter top 500 movies
top_500_movies = sorted_data.head(500)
```

```python
import scipy.stats

# Calculate mean, median, and mode for budget, revenue, and runtime
central_tendency = top_500_movies[['budget', 'revenue', 'runtime']].agg(['mean', 'median'])

# Mode calculation using scipy.stats.mode for each column
mode_budget = scipy.stats.mode(top_500_movies['budget']).mode[0]
mode_revenue = scipy.stats.mode(top_500_movies['revenue']).mode[0]
mode_runtime = scipy.stats.mode(top_500_movies['runtime']).mode[0]

# Add mode to the central_tendency DataFrame
central_tendency.loc['mode'] = [mode_budget, mode_revenue, mode_runtime]

print("Central Tendency for Budget, Revenue, and Runtime:")
print(central_tendency)
```

```python
plt.figure(figsize=(15, 5))

# Boxplot for Budget
plt.subplot(1, 3, 1)
sns.boxplot(x=top_500_movies['budget'])
plt.title('Budget')

# Boxplot for Revenue
plt.subplot(1, 3, 2)
sns.boxplot(x=top_500_movies['revenue'])
plt.title('Revenue')

# Boxplot for Runtime
plt.subplot(1, 3, 3)
sns.boxplot(x=top_500_movies['runtime'])
plt.title('Runtime')

# Show the plots
plt.show()
```
